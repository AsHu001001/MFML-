```python
import numpy as np
def g(A,b):
 A=np.array(A,float)
 b=np.array(b,float)
 n=A.shape[0]
 M=np.hstack((A,b.reshape(-1,1)))
 for k in range(n):
  mr=max(range(k,n),key=lambda x:abs(M[x,k]))
  if M[mr,k]==0:continue
  if mr!=k:
   M[[k,mr]]=M[[mr,k]]
  for i in range(k+1,n):
   f=M[i,k]/M[k,k]
   M[i,k:]=M[i,k:]-f*M[k,k:]
 x=np.zeros(n)
 for i in range(n-1,-1,-1):
  x[i]=(M[i,-1]-M[i,i+1:n].dot(x[i+1:n]))/M[i,i]
 return x

if __name__=="__main__":
 print(g([[2,1,-1],[-3,-1,2],[-2,1,2]],[8,-11,-3]))
    """[ 2.  3. -1.]"""
```
